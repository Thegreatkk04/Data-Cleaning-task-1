```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv", sep="\t")

print(df.head())
print(df.columns)

# Missing values
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Save cleaned data
df.to_csv("cleaned_marketing_campaign.csv", index=False)
```
